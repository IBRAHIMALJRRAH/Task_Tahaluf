<b>Task 1 : Image Transformation using OpenCV</b>

Using mr bean image code the following cells:

```python
#read image
import cv2
import matplotlib.pyplot as plt
bean = cv2.imread("/content/bean.png")
```

```python
plt.imshow(cv2.cvtColor(bean, cv2.COLOR_BGR2RGB))
plt.title('Original Image 1')
plt.axis('off')
plt.show()
```

Q1: Image Translation -> shift of an image from one location to another

```python
#start code
# Image Translation
import numpy as np

rows, cols, _ = bean.shape
tx, ty = 50, 30  # Define the translation in the x and y directions

# Define the translation matrix
translation_matrix = np.float32([[1, 0, tx], [0, 1, ty]])

# Apply the translation to the image
translated_bean = cv2.warpAffine(bean, translation_matrix, (cols, rows))

# Display the translated image
plt.imshow(cv2.cvtColor(translated_bean, cv2.COLOR_BGR2RGB))
plt.title('Translated Image')
plt.axis('off')
plt.show()

#end code
```

Q2: Image Reflection -> flip the image horizontally & vertically

```python
# Horizontal Flip
flipped_horizontal_bean = cv2.flip(bean, 1)

# Display the horizontally flipped image
plt.imshow(cv2.cvtColor(flipped_horizontal_bean, cv2.COLOR_BGR2RGB))
plt.title('Horizontally Flipped Image')
plt.axis('off')
plt.show()
```

```python
#start code
# Vertical Flip
flipped_vertical_bean = cv2.flip(bean, 0)

# Display the vertically flipped image
plt.imshow(cv2.cvtColor(flipped_vertical_bean, cv2.COLOR_BGR2RGB))
plt.title('Vertically Flipped Image')
plt.axis('off')
plt.show()
#end code
```

Q3: Image Rotation -> rotate the image in any degree you want eg: (30, 60, 90)

```python
#start code
# Rotation Angles
angles = [30, 60, 90]

# Perform rotation for each angle
for angle in angles:
    # Define the rotation matrix
    rotation_matrix = cv2.getRotationMatrix2D((cols/2, rows/2), angle, 1)

    # Apply rotation to the image
    rotated_bean = cv2.warpAffine(bean, rotation_matrix, (cols, rows))

    # Display the rotated image
    plt.imshow(cv2.cvtColor(rotated_bean, cv2.COLOR_BGR2RGB))
    plt.title(f'Rotated Image ({angle} degrees)')
    plt.axis('off')
    plt.show()
#end code
```

Q4: Image Scaling -> resize image in any height and width you want eg: (480,640), (720, 800)

```python
#start code
# Scaling Dimensions
dimensions = [(480, 640), (720, 800)]

# Perform scaling for each dimension
for dim in dimensions:
    # Resize the image
    scaled_bean = cv2.resize(bean, dim)

    # Display the scaled image
    plt.imshow(cv2.cvtColor(scaled_bean, cv2.COLOR_BGR2RGB))
    plt.title(f'Scaled Image ({dim[0]}x{dim[1]})')
    plt.axis('off')
    plt.show()
#end code
```

Q5: Image Cropping -> remove the background pixel of image (not importent)

```python
#start code
roi_coordinates = (100, 50, 400, 500)  # Adjust these values based on your image and region of interest

# Crop the image
cropped_bean = bean[roi_coordinates[1]:roi_coordinates[1] + roi_coordinates[3],
                    roi_coordinates[0]:roi_coordinates[0] + roi_coordinates[2]]

# Display the cropped image
plt.imshow(cv2.cvtColor(cropped_bean, cv2.COLOR_BGR2RGB))
plt.title('Cropped Image (Foreground)')
plt.axis('off')
plt.show()

#end code
```

Q6: Image Shearing -> shear the image once on X-Axis & Y-Axis

X-Axis Shearing:

```python
#start code
# X-Axis Shearing
shear_factor_x = 0.2  # Adjust the shear factor based on your preference

# Define the shear matrix for X-axis
shear_matrix_x = np.float32([[1, shear_factor_x, 0], [0, 1, 0]])

# Apply X-axis shearing to the image
sheared_bean_x = cv2.warpAffine(bean, shear_matrix_x, (cols, rows))

# Display the X-axis sheared image
plt.imshow(cv2.cvtColor(sheared_bean_x, cv2.COLOR_BGR2RGB))
plt.title('X-Axis Sheared Image')
plt.axis('off')
plt.show()
#end code
```

Y-Axis Shearing:

```python
# Y-Axis Shearing
shear_factor_y = 0.2  # Adjust the shear factor based on your preference

# Define the shear matrix for Y-axis
shear_matrix_y = np.float32([[1, 0, 0], [shear_factor_y, 1, 0]])

# Apply Y-axis shearing to the image
sheared_bean_y = cv2.warpAffine(bean, shear_matrix_y, (cols, rows))

# Display the Y-axis sheared image
plt.imshow(cv2.cvtColor(sheared_bean_y, cv2.COLOR_BGR2RGB))
plt.title('Y-Axis Sheared Image')
plt.axis('off')
plt.show()
```
